# 2. Créer son premier graphique

```python
# import de matplotlib pour la visualisation
import matplotlib.pyplot as plt

# Import de Numpy
import numpy as np

# Import sql fonctionnality for testing cause i'm profeciency at it
import pandasql as pdsql
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
# On initialise la figure
# On souhaite effacer le Out tout rouge on utilisera plt.draw()
black_friday.plot()
plt.draw()
```

```python
q = ''' Select sum(purchase) as Amound_total,Age from black_friday group by Age; '''
# Affichons notre requête
df = pysqldf(q)
df
```

```python
# DF plot
#df.plot(kind='bar')

# ou

# marker = 'o' est pour les points plus précis sur le graphique
plt.plot(df["Age"],df["Amound_total"], c='r', marker ='o')
```

```python
# On veut des point et non des ligne
plt.plot(df["Age"],df["Amound_total"], c='r', marker ='x', linestyle="")
```

# Ajouter un titre principal et des labels aux axes

```python
# Rentrons les données
plt.plot(df["Age"],df["Amound_total"], c='r', marker ='o', linestyle="")

# Titré le graphe (Y = 1.05 est pour replacé le titre qui se chevauche souvent avec le 1e9)
plt.title("Somme totale des ventes effectuer par catégorie d'âge", y = 1.05)

# Titrons l'axe x
plt.xlabel("Catégorie d'âge")

# Titrons l'axe y
plt.ylabel("Somme totale des ventes")

#Dessinons
plt.draw()
```

# Créer des graphiques avec Seaborn

```python
# Importer Seaborn
import seaborn as sns

# Lien pour les couleurs
# https://seaborn.pydata.org/generated/seaborn.color_palette.html
```

```python
plt.rcParams["figure.figsize"] = [9,9]
sns.boxplot("Gender", y="Purchase", data = black_friday, palette = "Set1")
plt.draw()
```

```python
# Selon quelle variable j'ai envie de colorier mon boxplot
# hue = "Variable"
sns.boxplot("Gender", y="Purchase", hue = "Gender", data = black_friday, palette = "Set1")
plt.draw()
```

```python
# Un box plot pour les femmes et un boxplot pour les hommes et selon la Catégorie d'âge
# hue = "Age"
# Loc = 2  --> met la légende en haut à gauche
# bbox_to_anchor = (1.05,1) ---> tu me la décale à droite de 1 et de 1
sns.boxplot("Gender", y="Purchase", hue = "Age", data = black_friday, palette = "Set1")
plt.legend(bbox_to_anchor = (1.05,1), loc = 2)
plt.draw()
```

```python
# Diagramme en batôn
sns.countplot(x="Age", hue ="Gender", data = black_friday, palette = "Set1")
```

```python
# Scatterplot
sns.catplot(x="Age", y = "Purchase", data = black_friday)
```

```python
subset = black_friday.sample(3000)
```

```python
subset
```

```python
sns.catplot(x="Age", y = "Purchase", data = subset)
plt.draw()
```

```python
# Pour que les point ne se chevauche pas
sns.catplot(x="Age", y = "Purchase", data = subset, kind="swarm")
plt.draw()
```

```python
# Colorer les point par sexe
# Plus d'homme ont fait des achats que des femmes
sns.catplot(x="Age", y = "Purchase", hue = "Gender" ,data = subset, kind="swarm")
plt.draw()
```

Tracer un barplot avec la fonction sns.barplot représentant la variable total_bill et en séparant ces données par sexe


```python
plt.rcParams["figure.figsize"] = [5,5]
sns.barplot(x="sex", y="total_bill", data = tips)
plt.draw()
```

Utiliser la fonction sns.distplot sur la variable total_bill pour étudier sa distribution


```python
sns.distplot(tips["total_bill"])
plt.draw()
```

```python
# Si on veut enlever la courbe
sns.distplot(tips["total_bill"], kde= False)
plt.draw()
```

Utiliser la fonction jointplot de Seaborn pour tracer la variable total_bill (X) en fonction de la variable tip (Y) 


```python
# Un graphique Bivarié, au milieu et univarié les deux au fond bleu
sns.jointplot(x="total_bill", y = "tip", data = tips)
plt.draw()
```

```python
sns.jointplot(x="total_bill", y = "tip", data = tips, kind = "reg")
plt.draw()
```

Utiliser la fonction pairplot de Seaborn sur l’ensemble du dataframe « tips » pour étudier les relations entre l’ensemble des variables.


```python
sns.pairplot(tips)
plt.draw()
```

Tracer le boxplot de la variable total_bill en créant un boxplot par jour (variable day) et en utilisant la variable sex dans l’option hue

```python
sns.boxplot(x="day", y="total_bill", hue = "sex", data = tips)
plt.draw()
```
